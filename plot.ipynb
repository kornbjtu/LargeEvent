## Batch Plot

### Seed independence

```python
import matplotlib.pyplot as plt
import csv
import pandas as pd
import seaborn as sns
import os

plt.style.use('default')
folder = "output"
file = lambda i: "output_scenario_1_seed_{0}.csv".format(str(i))
```

```python
l = []

for i in range(1, 101):
    data = pd.read_csv("/".join([folder, file(i)]))
    avg = data['order_number'].max()

    l.append(avg)



    
```

```python


# Set the aesthetic style of the plots

# Create the plot
plt.figure(figsize=(6,4))
histplot = sns.histplot(l, bins=50, kde=True, color="blue", line_kws={"linewidth": 2})

# Customize the plot
plt.xlabel("Number of Order", fontsize=14, fontweight='bold')
plt.ylabel("Density", fontsize=14, fontweight='bold')
plt.title("Histogram of Average Waiting time with Fitted KDE", fontsize=16, fontweight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
